```python
import yfinance as yf

# ดึงข้อมูลดัชนี SET50
set50 = yf.Ticker("^SET.BK")
data = set50.history(period="10y")

print(data.head())
```

```python
import pandas_ta as ta

# คำนวณ RSI, MACD, และ Bollinger Bands
data.ta.rsi(append=True)
data.ta.macd(append=True)
data.ta.bbands(append=True)

# ลบแถวที่มีข้อมูลไม่สมบูรณ์ (NaN) ที่เกิดจากการคำนวณ
data.dropna(inplace=True)

print(data.tail())
```

```python
# กำหนด path ที่จะบันทึกไฟล์
save_path = '/content/drive/MyDrive/Thai_Quant_AI_Project/02_data/set50_processed_data.csv'

# บันทึกเป็นไฟล์ CSV
data.to_csv(save_path)

print(f"ข้อมูลถูกบันทึกเรียบร้อยที่: {save_path}")
```
